#### Creating study dataset

```python
dataset = (
    cleaned_df2.
    merge(cleaned_df3, left_on=["Year", "Country"], right_on=["Year", "admin"], how="outer", suffixes=("_df2", "_df3")).
    merge(cleaned_df4, left_on=["Year", "Country"], right_on=["Year", "Country Name"], how="outer", suffixes=(None, "_df4")).
    merge(cleaned_df5, on=["Year", "Country"], how="outer", suffixes=(None, "_df5")).
    merge(cleaned_df6, on=["Year", "Country"], how="outer", suffixes=(None, "_df6")).
    merge(cleaned_df7, on=["Year", "Country"], how="outer", suffixes=(None, "_df7")).
    merge(cleaned_df8, on=["Year", "Country"], how="outer", suffixes=(None, "_df8")).
    merge(cleaned_df9, left_on=["Year", "Country"], right_on=["Year", "Country Name"], how="outer", suffixes=(None, "_df9")). 
    merge(vaccination_status_df, on=["Year", "Country"], how="outer")
)
```
